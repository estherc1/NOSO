```python
sale = pd.read_csv('sale_by_week.csv')
```

```python
sample = sale.loc[sale['Style_Color'] == 'DTAN1190CHA']
sample.head()
```

```python
del(sample['Style_Color'])
del(sample['Unnamed: 0'])
```

```python
sample.head()
```

```python
m = Prophet(changepoint_prior_scale=0.05).fit(sample)
```

```python
future = m.make_future_dataframe(periods=52, freq = 'W')
```

```python
forecast = m.predict(future)
forecast[['ds','yhat']]
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```

```python
metric_df = forecast.set_index('ds')[['yhat']].join(sample.set_index('ds').y).reset_index()
```

```python
metric_df.dropna(inplace=True)
```

```python
r2_score(metric_df.y, metric_df.yhat)
```

```python
 x = mean_squared_error(metric_df.y, metric_df.yhat)
math.sqrt(x)
```

```python
mean_absolute_error(metric_df.y, metric_df.yhat)
```
